```python
import pandas as pd
```

```python
dfx = pd.get_dummies(df['sex'])
# sex : female -> 0, male -> 1
dfx
```

```python
df = pd.concat([df,dfx],axis=1)
df
```

```python
dfx = pd.get_dummies(df['pclass'],prefix='pclass')
df = pd.concat([df,dfx],axis=1)
df
#pd.get_dummies(df['class'],prefix='pclass',drop_first=True)
# y(survive) = ax(pclass) + 0
```

female	male pclass_First	pclass_Second	pclass_Third

```python
df[['fare','female', 'male', 'pclass_1', 'pclass_2', 'pclass_3','survived']].head(5)
```

```python
X = df[['fare','female', 'male', 'pclass_1', 'pclass_2', 'pclass_3']]
Y = df[['survived']]
X.shape, Y.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
X.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y)
```

```python
from sklearn.linear_model import LogisticRegression
logR = LogisticRegression()
logR.fit(x_train, y_train)
```

```python
logR.score(x_train, y_train)
```
